```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
runs_dir = '/content/runs/detect'
latest_train = max([os.path.join(runs_dir, d) for d in os.listdir(runs_dir)], key=os.path.getmtime)
path_matrix = os.path.join(latest_train, 'confusion_matrix.png')
if os.path.exists(path_matrix):
    img = mpimg.imread(path_matrix)
    plt.figure(figsize=(5, 5))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
else:
    print("Coba cek folder runs/detect/train.")
```
